```python
import pandas as pd
from datasets import load_dataset 
import matplotlib.pyplot as plt

#loading data
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#cleaning data
df['job_posted_data']=pd.to_datetime(df['job_posted_date'])
```

```python
top_countries = df['job_country'].value_counts().head(6).index.tolist()  # Convert to list

df_job_country_salary = df.pivot_table(
    values='salary_year_avg', 
    index='job_title_short',
    columns='job_country',
    aggfunc='median'
)
df_job_country_salary = df_job_country_salary.loc[:, top_countries]  # Use : to select all rows and top_countries for columns

job_titles=['Data Analyst','Data Engineer','Data Scientist']


df_job_country_salary = df_job_country_salary.loc[job_titles]
df_job_country_salary.plot(kind='bar')
```
